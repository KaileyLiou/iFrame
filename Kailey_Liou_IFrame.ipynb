# Part 2: Working with Colors & Text

Here's how to create gradients with pixel manipulation:

````python
# Here's a colorful gradient example
width, height = 300, 300
img = Image.new('RGB', (width, height))

# This creates a simple gradient - try to understand how!
for x in range(width):
    for y in range(height):
        r = int((x / width) * 255)
        g = int((y / height) * 255)
        b = 100
        img.putpixel((x, y), (r, g, b))

plt.imshow(img)
plt.axis('off')
plt.show()
````

## Challenge 2: Custom Gradient ðŸŽ¯

**Your task:** Modify the gradient code to create a different color pattern. Try changing how `r`, `g`, and `b` are calculated.

*Hint: Try using math functions like `sin`, `cos`, or different formulas*

````python
# Create your own gradient here
# Hint: Try using math functions like sin, cos, or different formulas

import math

width, height = 300, 300
img = Image.new('RGB', (width, height))

for x in range(width):
    for y in range(height):
        r = int(math.sin(x / width) * 255)
        g = int(math.cos(y / height) * 255)
        b = 100
        img.putpixel((x, y), (r, g, b))

plt.imshow(img)
plt.axis('off')
plt.show()
````

## Advanced Image Editing

Here are some more advanced techniques to try:

````python
# Color channel manipulation
img_copy = original_img.copy().convert('RGB')
width, height = img_copy.size

# Example: Swap red and blue channels
pixels = list(img_copy.getdata())
swapped_pixels = []

for pixel in pixels:
    r, g, b = pixel
    # Swap red and blue
    swapped_pixels.append((b, g, r))

img_copy.putdata(swapped_pixels)

# Show before and after
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.imshow(original_img)
ax1.set_title('Original')
ax1.axis('off')

ax2.imshow(img_copy)
ax2.set_title('Red-Blue Swapped')
ax2.axis('off')

plt.show()
````
